```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
stocks = pd.read_csv('../data/data_visualization/stocks.csv')
stocks.head()
```

```python
stocks['Date'] = pd.to_datetime(stocks['Date'])
print(stocks.Date.dtype)
stocks.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax = sns.lineplot(x="Date", y="AAPL", data=stocks)
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))

ax1 = fig.add_subplot(231)
ax1.plot(stocks.Date, stocks.AAPL, color="g")
ax1.set_title("AAPL")

ax2 = fig.add_subplot(232)
ax2.plot(stocks.Date, stocks.ADBE, color="r")
ax2.set_title("ADBE")

ax3 = fig.add_subplot(233)
ax3.plot(stocks.Date, stocks.GOOG, color="b")
ax3.set_title("GOOG")

ax4 = fig.add_subplot(234)
ax4.plot(stocks.Date, stocks.NFLX, color="magenta")
ax4.set_title("NFLX")

ax5 = fig.add_subplot(235)
ax5.plot(stocks.Date, stocks.IBM, color="orange")
ax5.set_title("IBM")

ax6 = fig.add_subplot(235)
ax6.plot(stocks.Date, stocks.ORCL, color="teal")
ax6.set_title("ORCL")

plt.show()
```
