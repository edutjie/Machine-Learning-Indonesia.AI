# Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('../data/data_visualization/iris.csv')
iris.head()
```

```python
iris.species.unique()
```

```python
iris.isnull().any()
```

```python
iris.describe()
```

```python
iris.species.hist()
plt.show()
```

```python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

No Correlation!

```python
iris.plot(kind='scatter', x='petal_length', y='petal_width')
plt.show()
```

Almost Perfect Correlation!

```python
sns.distplot(iris.sepal_length)
plt.show()
```

The data distributed quite ideal! and no outlier but there's 2 hill.

```python
sns.distplot(iris.sepal_width)
plt.show()
```

This data distributed better and very ideal!

```python
sns.boxplot(data=iris)
plt.show()
```

But there we can see that sepal_width feature has few outliers

```python
sns.pairplot(data=iris, hue='species')
plt.show()
```

```python
correlation = iris.corr()
correlation
```

```python
sns.heatmap(correlation, annot=True)
plt.show()
```
