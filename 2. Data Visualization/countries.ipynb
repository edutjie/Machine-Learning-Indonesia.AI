```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
countries = pd.read_csv('../data/data_visualization/countries.csv')
countries.head()
```

```python
dt_2007 = countries[countries['year'] == 2007]
dt_2007
```

```python
top_10 = dt_2007.sort_values('population', ascending=False).head(10)
top_10
```

```python
fig, ax = plt.subplots(figsize=(15, 8))

r = range(10)
plt.bar(r, top_10.population)
plt.xticks(r, top_10.country, rotation=90)
plt.show()
```
